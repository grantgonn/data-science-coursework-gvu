## Exercise 3(a) (5 points)

Write a Python function to calculate the take rate using Equation 9-2.

```python
import numpy as np

def take_rate(price, beta_0 = 3.65, beta_1 = -0.55):
    Z = beta_0 + beta_1 * price
    return np.exp(Z) / (1 + np.exp(Z))
```

## Exercise 3(c) (5 points)

Simulate contribution using: Revenue = Take Rate × Market Size × Price, and Cost = $1.50 per subscriber.

```python
def contribution(price, market_size = 1e7, cost = 1.5):
    tr = take_rate(price)
    revenue = price * tr * market_size
    cost_total = cost * market_size * tr
    return revenue - cost_total

# simulate the contribution for a range of prices
prices = np.arange(0, 10, 0.01)
contributions = contribution(prices)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(prices, contributions, label='Contribution')
plt.title('Contribution vs Price')
plt.xlabel('Price')
plt.ylabel('Contribution')
plt.axhline(0, color='red', linestyle='--', label='Zero Contribution')
plt.legend() 
```
