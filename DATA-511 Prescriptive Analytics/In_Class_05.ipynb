# Exercise 2 (5 points)

Compute the probability of getting exactly 2 heads in 3 tosses of a fair coin.

Using $\text{Bin}(n=3,p=0.5)$ \
$P(X=2)=\binom{3}{2}(0.5)^2(0.5)^{3-2}=3×0.25×0.5=0.375$

```python
from scipy.stats import binom

prob = binom.pmf(2, 3, 0.5)
print(f"The probability of getting exactly 2 heads in 3 flips of a fair coin is: {prob:.4f}")
```

# Exercise 4 (5 points)

For the standard normal distribution N(0, 1), what is the probability that a value falls between -1.3 and 1.7?

```python
# X ~ N(0, 1) we want to find P(-1.3 < X < 1.7)

from scipy.stats import norm

probability = norm.cdf(1.7) - norm.cdf(-1.3)
print(f"The probability that X is between -1.3 and 1.7 is: {probability:.4f}")
```

# Exercise 5 (10 points)

A company manufactures two products: Product A and Product B. Each unit of Product A requires 2 hours of labor and 3 units of raw material. Each unit of Product B requires 4 hours of labor and 2 units of raw material. The company has a total of 100 labor hours and 90 units of raw material available. The profit for each unit sold is $40 for Product A and $50 for Product B. Find the optimal number of Products A and B that maximizes company's profit.

```python
# the goal is to maximize profit -> profit = 40x +50y (x = product A, y = product B)
# constraints: x >= 0, y >= 0, 
# 2x +4y <= 100
# 3x + 2y <= 90

from scipy.optimize import linprog
c = [-40, -50]  # Coefficients for the objective function (negative for maximization)

# define the inequality constraints matrix
A = [[2, 4], [3, 2]]
b = [100, 90] 

# bounds for x and y
x_bounds = (0, None)
y_bounds = (0, None)

# solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])

if result.success:
    print(f"Optimal number of product A: {result.x[0]:.2f}")
    print(f"Optimal number of product B: {result.x[1]:.2f}")
    print(f"Maximum profit: {-result.fun:.2f}")
else:
    print("Optimization failed")
```
