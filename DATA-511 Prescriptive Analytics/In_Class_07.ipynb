# Exercise 4 (10 points)

You are given three development scenarios for a hotel project with associated prior probabilities and annual earnings over a 10-year horizon. The discount rate is 5%. Your task is to compute the expected present value of the project using the formula for an ordinary annuity. The data is shown below:

```
scenarios = ["Low", "Medium", "High"]
probabilities = [0.2, 0.5, 0.3]
earnings = [1.1, 1.6, 2.0]  # in billions
discount_rate = 0.05
term = 10
```

Write a Python function to compute the present value of an ordinary annuity and use it to calculate the expected present value of the project based on the scenarios above.

````python
scenarios = ["Low", "Medium", "High"]
probabilities = [0.2, 0.5, 0.3]
earnings = [1.1, 1.6, 2.0]  # in billions
discount_rate = 0.05
term = 10

def present_value_annuity(earnings, r, t):
    """Calculate the present value of an annuity."""
    pv = (earnings/ r) * (1 - 1/(1 + r) ** t)
    return pv
````

````python
# calculate the expected present value of each scenario
expected_pv = [present_value_annuity(e, discount_rate, term) for e in earnings]
print("Expected Present Value of each scenario:", expected_pv)

# compute the expected present value
expected_value = sum(p * pv for p, pv in zip(probabilities, expected_pv))
print("Expected Present Value of the project:", expected_value)
````
