# Exercise 3 (5 points)

Write a Python function to simulate tossing a coin 10,000 times. Use 0 for tails and 1 for heads.

```python
import numpy as np 

def sim_toss(nsim):
    results = np.random.choice([0,1], p=[0.5,0.5], size=nsim)
    return results

# Simulate 10000 coin tosses
nsim = 10000
results = sim_toss(nsim) 
print("proportion of heads:", np.mean(results))
```

# Exercise 6 (7 points)

Using Python, simulate 1000 regressions and store coefficients of each run.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# we assume y = 50 +25*x + error
# assume x is uniform between .5 and .8

def lr_sim(nsim):
    slope = []
    for _ in range(nsim):
        # Simulate x and y
        x = np.random.uniform(0.5, 0.8, nsim)
        error = np.random.normal(0, 2, nsim)  
        y = 50 + 25 * x + error
        df = pd.DataFrame({'y': y, 'x': x})
        model = smf.ols('y ~ x', data=df).fit()
        slope.append(model.params['x'])
    return slope

nsim = 1000
slopes = lr_sim(nsim)

plt.hist(slopes, bins=30, edgecolor='black')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.title('Distribution of Slopes from Simulated Linear Regression')
plt.axvline(x=25, color='red', linestyle='--', label='True Slope')
plt.legend()
plt.show()
```
