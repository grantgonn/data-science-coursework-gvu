# Exercise 4 (10 points)

A pastry shop produces two types of pastries: Croissants (X) and Danishes (Y). Each pastry requires:

| Resource | Croissant (X) | Danish (Y) | Available | 
| --- | --- | --- | --- |
| Flour (grams) | 200 | 100 | 5000 |
| Labor (minutes) | 10 | 15 | 400 |
| Profit per unit | $2 | $3 | |

The owner wants to maximize profit while not exceeding available resources. Formulate and solve this as a linear programming problem using Python.

```python
# Maximize 2X + 3Y
# subject to: 200X + 100Y <= 5000 (flour constraint)
#            10X + 15Y <= 400 (labor constraint)
#             X >= 0, Y >= 0 (non-negativity constraints)

from scipy.optimize import linprog
# coefficents of the objective function to maximize
c = [-2, -3]  # Negate for maximization

# Coefficients for the inequality constraints
A = [[200, 100], [10, 15]]

# Right-hand side of the inequality constraints
b = [5000, 400]

# Bounds for each variable (X and Y)
x_bound = (0, None)  # X >= 0
y_bound = (0, None)  # Y >= 0

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bound, y_bound])

# Check if the optimization was successful
if result.success:
    print("Optimal value of X:", result.x[0])
    print("Optimal value of Y:", result.x[1])
    print("Maximum value of the objective function: $", -result.fun)  # Negate to get the max value
else:
    print("Optimization failed:", result.message)
```

# Exercise 5 (10 points)

Consider the context presented in Exercise 4. This time solve the program as an interger linear programming. Formulate and solve the problem as an integer programming problem in Python.

```python
import numpy as np
from scipy.optimize import milp, LinearConstraint, Bounds

# Define the coefficients of the objective function
c_milp = [-2, -3] 

# Define the inequality constraints
A_milp = [[200, 100], [10, 15]]
# Right-hand side of the inequality constraints
b_milp = [5000, 400]

# set up the linear constraints
linear_constraint = LinearConstraint(A_milp, lb=-np.inf, ub=b_milp)

# Define the bounds for each variable
bounds_milp = Bounds([0, 0], [np.inf, np.inf]) 

# Interger constraints
intergrality = [1, 1]  

# Solve the mixed-integer linear programming problem
result_milp = milp(c_milp, constraints=[linear_constraint], bounds=bounds_milp, integrality=intergrality)

# Check if the optimization was successful
if result_milp.success:
    print("Optimal integer value of X:", result_milp.x[0])
    print("Optimal integer value of Y:", result_milp.x[1])
    print("Maximum integer value of the objective function: $", -result_milp.fun)  # Negate to get the max value
else:
    print("Mixed-integer optimization failed:", result_milp.message)
```
