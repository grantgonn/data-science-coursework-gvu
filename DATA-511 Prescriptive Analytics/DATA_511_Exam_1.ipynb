# Exercise 7 (4 points)

Suppose $X \sim N(100, 5)$. What is the probability that $X$ is less than 110? Use Python.

```python
from scipy.stats import norm

mean = 100
std = 5

prob = norm.cdf(110, loc=mean, scale=std)
print(f"The probability of a value being less than or equal to 110 is: {prob:.4f}")
```

# Exercise 14 (10 points)

A manufacturer produces two products: Product A and Product B. Each unit of Product A requires 2 hours of machining and 3 hours of labor. Each unit of Product B requires 4 hours of machining and 2 hours of labor. The company has a maximum of 100 machining hours and 90 labor hours available per week. The profit is $30 per unit of Product A and $50 per unit of Product B. Formulate and solve the problem as linear programming in Python

```python
from scipy.optimize import linprog

# maximize profit = 30x + 50y
# 2x + 4y <= 100
# 3x + 2y <= 90
# x >= 0, y >= 0

# Coefficients for the objective function
c = [-30, -50]  

# Coefficients for the inequality constraints
A = [[2, 4], [3, 2]]
b = [100, 90]

# Bounds for each variable
x_bounds = (0, None)  # x >= 0
y_bounds = (0, None)  # y >= 0

result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])

if result.success:
    print(f"Optimal number of product A: {result.x[0]:.2f}")
    print(f"Optimal number of product B: {result.x[1]:.2f}")
    print(f"Maximum profit: {-result.fun:.2f}")
else:
    print("Optimization failed")
```
