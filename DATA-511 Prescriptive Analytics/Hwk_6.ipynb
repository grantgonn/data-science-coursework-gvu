A company is choosing between two investment locations: City Center and Suburb. Each location has three development options (No Development, Modest, Major), with associated prior probabilities and expected annual earnings. The investment horizon is 10 years, and the discount rate is 6%.

```
import pandas as pd

data = {
    "Location": ["City Center"] * 3 + ["Suburb"] * 3,
    "Development": ["No Dev", "Modest", "Major"] * 2,
    "Prior": [0.2, 0.3, 0.5, 0.3, 0.4, 0.3],
    "Annual_Earnings": [1.0, 1.4, 2.0, 1.1, 1.6, 1.9],
}

discount_rate = 0.06
term = 10

```

## Exercise 7(a) (5 points)

Calculate the present value of earnings for each scenario.

````python
import pandas as pd

data = {
    "Location": ["City Center"] * 3 + ["Suburb"] * 3,
    "Development": ["No Dev", "Modest", "Major"] * 2,
    "Prior": [0.2, 0.3, 0.5, 0.3, 0.4, 0.3],
    "Annual_Earnings": [1.0, 1.4, 2.0, 1.1, 1.6, 1.9],
}

discount_rate = 0.06
term = 10

df = pd.DataFrame(data)

# Calculate present value of annuity
annuity_factor = (1 - (1 + discount_rate) ** -term) / discount_rate
df["PV_Earnings"] = df["Annual_Earnings"] * annuity_factor

df
````

## Exercise 7(b) (3 points)

Compute the expected present value per development type.

````python
df["Weighted_PV"] = df["Prior"] * df["PV_Earnings"]
expected_pv_by_dev = df.groupby("Development")["Weighted_PV"].sum()

expected_pv_by_dev
````

## Exercise 7(c) (3 points)

Compute the total expected present value per location.

````python
df["Weighted_PV"] = df["Prior"] * df["PV_Earnings"]
expected_pv_by_location = df.groupby("Location")["Weighted_PV"].sum()

expected_pv_by_location
````
