# Exercise 2 (5 points)

Given five executives (CEO, CFO, COO, CMO, CLO), how many unique pairs can be formed for interviews?

(5 2) = 5! / 2!(5-2)! = 10

# Exercise 4 (4 points)

Given the events A and B are mutually exclusive, with $P(A) = 0.3$ and $P(B) = 0.5$, find $P(A \cup B)$.

P(A U B) = .3 + .5 + 0 = .8

# Exercise 5 (5 points)

In a survey, 150 out of 300 people are male, and 100 of these males like a product. What is the conditional probability that a randomly selected male likes the product?

P(likes product | male) = 100/150 = 2/3

# Exercise 7 (6 points)

Define the sample space for a fair coin tossed twice. List all possible outcomes. What is the probability of getting exactly one head?

Sample space: {HH, HT, TH, TT}

P(One head) = {HT, TH} / {HH, HT, TH, TT} = 2/4 = .5
