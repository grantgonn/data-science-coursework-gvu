Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials that have the same probability of success. An example would be a multiple choice quiz with the same number of options and guessing for each one.

# Exercise 4 (5 points)

Suppose $X \sim N(100, 25)$. What is the probability that $X$ is less than 110?

```python
from scipy.stats import norm

mean = 100
std = 5

# Calculate P(X < 110)
x = 110
prob = norm.cdf(x, loc=mean, scale=std)

print(f"P(X < {x}) = {prob:.4f}")
```

# Exercise 5 (10 points)

A farmer has 80 acres of land available for planting wheat and corn. Each acre of wheat yields a profit of $200 and requires 3 hours of labor. Each acre of corn yields a profit of $300 and requires 4 hours of labor. The farmer has at most 240 labor hours available during the planting season. Find the optimal number of acres to plant wheat and corn that maximizes the farmers profit.

```python
# the goal is to maximize profit -> profit = 200x + 300y (x = Wheat, y = Corn)
# constraints: x >= 0, y >= 0, 
# 3x + 4y <= 240
# x + y <= 80

from scipy.optimize import linprog
c = [-200, -300]  # Coefficients for the objective function (negative for maximization)

# define the inequality constraints matrix
A = [[3, 4], [1, 1]]
b = [240, 80] 

# bounds for x and y
x_bounds = (0, None)
y_bounds = (0, None)

# solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])

if result.success:
    print(f"Optimal number of acres for wheat: {result.x[0]:.2f}")
    print(f"Optimal number of acres for corn: {result.x[1]:.2f}")
    print(f"Maximum profit: {-result.fun:.2f}")
else:
    print("Optimization failed")
```
