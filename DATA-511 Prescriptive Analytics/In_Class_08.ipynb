# Exercise 6 (5 points)

Write a Python function to implement a linear congruential generator. Return a list of 5 values.

```python
def lcg(seed, a, c, m, n):
    """
    Generate a sequence of pseudo-random numbers using a Linear Congruential Generator (LCG).

    :param seed: Initial seed value
    :param a: Multiplier
    :param c: Increment
    :param m: Modulus
    :param n: Number of random numbers to generate
    :return: List of generated random numbers
    """
    random_numbers = []
    current = seed
    for _ in range(n):
        current = (a * current + c) % m
        random_numbers.append(current)
    return random_numbers
```
