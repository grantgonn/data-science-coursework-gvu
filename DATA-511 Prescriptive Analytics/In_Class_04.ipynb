# Exercise 5 (6 points)

A survey shows that 51.3% of those who like a new product are male. If 65% of respondents like the product, what is the probability that a randomly selected male likes the product (using Bayes’ Theorem)?

P(male|like) = 0.513

P(like) = .65

By Bayes' thorem: P(like|male) = P(like and male)/P(male)

(P(male|like) * P(like)) / P(male)

(.513 * .65) / P(male)

P(male) = P(male|like) * P(like) + P(male|don't like) * P(don't like) = .513 * .65 + (1-.513) * (1-.65) = .5

P(like|male) = (.513 * .65) / .5 = .667
