### Exercise 1(f) (10 points)

Use `Word2Vec` from `gensim` library to create embedding for `description_clean_2`. Use the following configuration for `Word2Vec`:

```
Word2Vec(sentences=tokenized_texts, vector_size=200, window=5, min_count=1, workers=4)
```

Store the embedding in a matrix called `X`.

````python
import gensim

from gensim.models import Word2Vec

tokens = wine_df['description_clean_2'].apply(lambda x: x.split())

# train word2vec
word2vec_model = Word2Vec(sentences=tokens, vector_size=200, window=5, min_count=1, workers=4)

# embed text
def get_avg_word2vec(token_list, model, vector_size):
    #remove out-of-vocab words
    valid_tokens = [word for word in token_list if word in model.wv]
    
    if not valid_tokens:
        return np.zeros(vector_size)
    
    return np.mean(model.wv[valid_tokens], axis=0)

# apply function to get embeddings
X = tokens.apply(lambda x: get_avg_word2vec(x, word2vec_model, 100))
X = np.vstack(X.values)
print(X[:5])
````

### Exercise 1(g) (10 points)

Build a neural network as follows:

- Build a neural network as follows:
   - `BatchNormalization()` layer.
   - `Dense` layer with 64 neurons and `activation=mish`.
   - `Dropout(0.1)`
   - `Dense` with 1 neuron `activation=None`.
- Use `optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)`, and `loss='MeanAbsoluteError'`.
- Train the network over 5 folds.
- Use `epochs=10` and `batch_size=32`. 
- Report the 5-fold mean absolute error. 

Use `X` as the inpute and `wine_df[points]` as the target variable.

````python
from sklearn.model_selection import KFold as kf
from sklearn.metrics import mean_absolute_error

y = wine_df['points']

md1 = tf.keras.models.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='mish', input_shape= (X.shape,)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(1)
])

md1.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), loss = 'MeanAbsoluteError')

# k fold
kfold = kf(n_splits=5, shuffle=True)
scores = []

for i, (train_idx, test_idx) in enumerate(kfold.split(X,y)):
    X_train, x_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    md1.fit(X_train, y_train, epochs = 10, batch_size = 32, verbose=0)
    
    y_pred = md1.predict(x_test, batch_size=32, verbose=False)
    MAE = mean_absolute_error(y_pred, y_test)
    scores.append(MAE)
    print(f'Fold {i+1}, {MAE}')
    
print('average MAE', np.mean(scores))
````

````python
md2 = tf.keras.models.Sequential([
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64, activation='mish', input_shape= (X.shape,)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='mish'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(1)
])

md2.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), loss = 'MeanAbsoluteError')

# k fold
kfold = kf(n_splits=5, shuffle=True)
scores = []

for i, (train_idx, test_idx) in enumerate(kfold.split(X,y)):
    X_train, x_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    md2.fit(X_train, y_train, epochs = 10, batch_size = 32, verbose=0)
    
    y_pred = md2.predict(x_test, batch_size=32, verbose=False)
    MAE = mean_absolute_error(y_pred, y_test)
    scores.append(MAE)
    print(f'Fold {i+1}, {MAE}')
    
print('average MAE', np.mean(scores))
````
