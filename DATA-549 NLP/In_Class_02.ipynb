### Exercise 1 (2 points)

Create an object called `text` with the following sentence:

```
Ethics are built right into the ideals and objectives of the United Nations
```

````python
text = 'Ethics are built right into the ideals and objectives of the United Nations'
text
````

### Exercise 2 (2 points)

Report the length of `text`.

````python
len(text)
````

### Exercise 3 (2 points)

Split `text` based on a single space and store the words in an object called `text_1`.

````python
text_1 = text.split(" ")
text_1
````

### Exercise 4 (2 points)

Report the length of `text_1`.

````python
len(text_1)
````

### Exercise 5 (2 points)

Report the words with more than 3 characters in `text_1`.

````python
[w for w in text_1 if len(w) > 3]
````

### Exercise 6 (2 points)

Report the capitalized words from `text_1`.

````python
[w for w in text_1 if w.istitle()]
````

### Exercise 7 (2 points)

Report the words that end in `s` from `text_1`.

````python
[w for w in text_1 if w.endswith('s')]
````

### Exercise 8 (3 points)

Using `set()` report the number of unique words from the following text:

```
To be or not to be
```

````python
text_2 = "To be or not to be"
text_2 = text_2.split(" ")
text_2
````

````python
set(text_2)
````

````python
len(set(text_2))
````

### Exercise 9 (3 points)

Convert the words from part 8 to lowercase and report the unique words using `set()`.

````python
text_3 = [W.lower() for W in text_2]
text_3
````

````python
set(text_3)
````
