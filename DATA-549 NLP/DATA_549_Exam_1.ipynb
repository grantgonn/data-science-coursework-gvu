You are given the following list of customer reviews:

```
reviews = [
    "The product is EXCELLENT!!! Very happy with the purchase.",
    "Worst service ever... I want my money back! #refund",
    "Good value for the price. Would recommend to anyone looking for affordable products.",
    "Mediocre quality, but fast delivery. Satisfied overall.",

Write a Python function called `clean_reviews` that:

Hint: see in-class assignment 6 for reference. The answer should be:

```
['product excellent happy purchase',
 'bad service ever want money back refund',
 'good value price would recommend anyone look affordable product',
 'mediocre quality fast delivery satisfied overall',
 'great product buy definitely']
```

````python
import re
import string
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# download NLTK resources
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

reviews = [
    "The product is EXCELLENT!!! Very happy with the purchase.",
    "Worst service ever... I want my money back! #refund",
    "Good value for the price. Would recommend to anyone looking for affordable products.",
    "Mediocre quality, but fast delivery. Satisfied overall.",
    "Great product! Will buy again, definitely."
]

# funtion to get the part of speech tag for lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# function to remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

pronouns = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves']

def clean_reviews(reviews):
    cleaned_reviews = []
    for review in reviews:
        # text to lower
        review = review.lower()
        # remove punctuation and special characters
        review = ''.join([char for char in review if char not in string.punctuation])
        review = ''.join([char for char in review if char not in '“”‘’'])
        # remove numbers
        review = remove_numbers(review)
        # tokenize
        words = word_tokenize(review)
        # remove stop words and pronouns
        words = [word for word in words if word not in stopwords and word not in pronouns]
        # perform lemmatization
        lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
        lemmatized_words = ' '.join(lemmatized_words)
        cleaned_reviews.append(lemmatized_words)
    return cleaned_reviews

cleaned_reviews = clean_reviews(reviews)
print(cleaned_reviews)
````

````python
import string
df['Description_clean'] = df['Description'].str.replace(f"[{string.punctuation}]", "", regex=True)
df.head()
````

````python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
pronouns = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves']

def clean_text(text):
    if not isinstance(text, str):
        return ""
    tokens = word_tokenize(text.lower())
    filtered_tokens = [word for word in tokens if word not in stop_words and word not in pronouns]
    return ' '.join(filtered_tokens)

df['Description_clean'] = df['Description_clean'].apply(clean_text)
df.head()
````

````python
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# download NLTK resources
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# funtion to get the part of speech tag for lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_text(text):
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
    return ' '.join(lemmatized_words)

df['Description_clean'] = df['Description_clean'].apply(lemmatize_text)
df.head()
````

````python
# function to remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

df['Description_clean'] = df['Description_clean'].apply(remove_numbers)
df.head()
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tfidf_vectorizer = TfidfVectorizer(
    max_features=100,
    max_df=0.8,  # Remove words that appear in more than 80% of documents
    min_df=5,    # Remove words that appear in fewer than 5 documents
    stop_words='english'
)

# Fit and transform the cleaned text data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Description_clean'])

# Initialize LDA with 4 topics
n_topics = 4
lda_model = LatentDirichletAllocation(
    n_components=n_topics,
    random_state=42,
    max_iter=100,
    learning_method='batch'
)

# Fit the LDA model to the TF-IDF matrix
lda_topic_features = lda_model.fit_transform(tfidf_matrix)
````

````python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
        
no_top_words = 10
tfidf_feature_names = tfidf_vectorizer.get_feature_names_out()
display_topics(lda_model, tfidf_feature_names, no_top_words)
````

## Exercise 11(f) (10 points)

Generate topic features using K-means clustering on BERT embeddings. This approach:
1. Uses BERT to create dense semantic embeddings of the text
2. Applies K-means clustering to group similar documents (consider 4 clusters)
3. Creates features based on cluster assignments and distances

````python
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer

print("loading BERT model...")
bert_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded.")

# Generate BERT embeddings for the cleaned text
print("Generating BERT embeddings...")
bert_embeddings = bert_model.encode(df['Description_clean'].tolist(), show_progress_bar=True, batch_size=32)

print(f'BERT embeddings shape: {bert_embeddings.shape}')
print(f'each document is represented by a {bert_embeddings.shape[1]}-dimensional vector')
````

````python
import numpy as np
from sklearn.preprocessing import normalize

final_kmeans = KMeans(n_clusters=4, random_state=42, n_init=20)
cluster_labels = final_kmeans.fit_predict(normalize(bert_embeddings))

# feature 1: one-hot encoding of cluster labels
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False)
cluster_labels_reshaped = cluster_labels.reshape(-1, 1)
cluster_onehot = onehot_encoder.fit_transform(cluster_labels_reshaped)

# feature 2: distance to cluster centroids
distances_to_centroids = final_kmeans.transform(normalize(bert_embeddings))

# feature 3: sofrmax probabilities based on distances
cluster_similarities = 1/(1+distances_to_centroids)
cluster_probabilities = cluster_similarities / cluster_similarities.sum(axis=1, keepdims=True)

print(f"cluster assignments shape: {cluster_onehot.shape}")
print(f"distances to centroids shape: {distances_to_centroids.shape}")
print(f"cluster probabilities shape: {cluster_probabilities.shape}")

# Combine all clustering features
X_cluster_features = np.hstack([cluster_onehot, distances_to_centroids, cluster_probabilities])
print(f"Combined clustering features shape: {X_cluster_features.shape}")
````

# Exercise 11(g) (15 points)

Train and evaluate three `LogisticRegression` models as follows:

- Using LDA features as input.
- Using NMF features as input.
- Combine LDA + NMF features as input.
- Bert cluster features.
- Bert embeddings.

Evaluate the five models over a five-fold cross validation strategy. Use the `accuracy` as measure of performance. Repor the performance of each of the models. What model would you use to predict `Label`?

````python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

y = df['Label'].map({'Household': 0, 'Books': 1, 'Electronics': 2, 'Clothing & Accessories': 3})
                     
X_1 = lda_topic_features
X_2 = nmf_topic_features
X_3 = np.hstack([X_1, X_2])  # Combine LDA and NMF features

print("Comparing model performance with different feature sets:")
print("==" * 50)

feature_sets = {
    'LDA': X_1,
    'NMF': X_2,
    'Combined': X_3,
    'BERT cluster features': X_cluster_features,
    'BERT embeddings': bert_embeddings
}

# Stratified Group K-Fold Cross-Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=200))

results = {}
for name, X in feature_sets.items():
    scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy', n_jobs=-1)
    results[name] = scores
    print(f"{name} Features: Mean Accuracy = {scores.mean():.4f}, Std = {scores.std():.4f}")
    print(f"Individual Fold Accuracies: {scores}")
    
# Identify the best feature set
best_feature_set = max(results, key=lambda k: results[k].mean())
print(f"\n Best Feature Set: {best_feature_set} with Mean Accuracy = {results[best_feature_set].mean():.4f}")
````
