# Exercise 1 

In this exercise, we will work with the `spam.csv` data file. This file contains text and a label attached to each text. The goal is to clean the text data and then build classifcation models. 

### Exercise 1(a) (2 points)

Load the `pandas` and `numpy` libraries.

```python
import pandas as pd
import numpy as np
```

### Exercise 1(b) (2 points)

Read the `csv` file and create a data-frame called `spam_df`.

```python
spam_df = pd.read_csv('spam.csv')
spam_df.head()
```

### Exercise 1(c) (2 points)

Using the `string` library, remove all the punctuation of the `text` column.

```python
import string

spam_df['text_clean'] = spam_df['text'].str.replace(f"[{string.punctuation}]", "", regex=True)
spam_df.head()
```

### Exercise 1(d) (10 points)

Using the `nltk` library, remove stopwords and pronouns from `text_clean` and store in another column called `'text_clean_1'`. 

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
pronouns = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves']

def clean_text(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [word for word in tokens if word not in stop_words and word not in pronouns]
    return ' '.join(filtered_tokens)

spam_df['text_clean_1'] = spam_df['text_clean'].apply(clean_text)
spam_df.head()
```

### Exercise 1(e) (10 points)

Perform Lemmatization of `text_clean_1` and store the results in another column called `text_clean_2` in the `spam_df` data frame.

```python
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# download NLTK resources
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# funtion to get the part of speech tag for lemmatization
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_text(text):
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
    return ' '.join(lemmatized_words)

spam_df['text_clean_2'] = spam_df['text_clean_1'].apply(lemmatize_text)
spam_df.head()
```

### Exercise 1(f) (5 points)

Based on the results from part 1(f), we see digits in some of the text in `text_clean_2`. Remove all the digits from `text_clean_2` and store the results in another column `text_clean_3` in the `spam_df` data frame.

```python
import re

# function to remove numbers
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

spam_df['text_clean_3'] = spam_df['text_clean_2'].apply(remove_numbers)
spam_df.head()
```

### Exercise 1(g) (5 points)

Generate topic features using BERT embeddings on `text_clean_3`.

```python
from sentence_transformers import SentenceTransformer

print("loading BERT model...")
bert_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded.")

# Generate BERT embeddings for the cleaned text
print("Generating BERT embeddings...")
bert_embeddings = bert_model.encode(spam_df['text_clean_3'].tolist(), show_progress_bar=True, batch_size=32)

print(f'BERT embeddings shape: {bert_embeddings.shape}')
print(f'each document is represented by a {bert_embeddings.shape[1]}-dimensional vector')
```

### Exercise 1(h) (5 points)

Apply the K-Means clustering algorithm to the bert embedding generated in the previous step.

1. Set the number of clusters to k=3.

2. Assign the cluster labels to a new column in `spam_df` called `'kmeans_cluster_label'`.

3. Generate new features by computing the distance of each document to the centroid of each of the k=3 clusters. Store these distances in a NumPy array called `X_kmeans_features`.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Defining the KMeans model
kmeans_md = KMeans(n_clusters=3, n_init=20, random_state=42)

# Scaling the embeddings before clustering
scaler = StandardScaler()
bert_embeddings_scaled = scaler.fit_transform(bert_embeddings)

# Fitting the KMeans model
kmeans_md.fit(bert_embeddings_scaled)

# Getting cluster labels
cluster_labels = kmeans_md.labels_

# Assign cluster labels to original data
spam_df['cluster'] = cluster_labels
print(spam_df[['text', 'cluster']].head(10))

# Computing the distance of each point to each cluster center
X_kmeans_features = kmeans_md.transform(bert_embeddings_scaled)

print(f"KMeans feature shape: {X_kmeans_features.shape}")
```

### Exercise 1(i) (5 points)

Hierarchical clustering is a type of density-based grouping used to organize documents based on similarity.

1. Apply Agglomerative Hierarchical Clustering to the L2​-normalized BERT embeddings (bert_embeddings) with k=3 clusters, using the `'ward'` linkage method.

2. Store the resulting cluster assignments in a new column called `'bert_hierarchical_label'` in `spam_df`.

3. Generate new features by computing the distance of each document to the centroid of each of the k=3 clusters. Store these distances in a NumPy array called `X_hierarchical_features`.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.pairwise import euclidean_distances

# Defining the Agglomerative Clustering model
agg_clust = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Fitting the Agglomerative Clustering model
agg_clust.fit(bert_embeddings_scaled)

# Getting cluster labels
spam_df['agg_cluster'] = agg_clust.labels_
print(spam_df[['text', 'agg_cluster']].head(10))

# Computing the distance of each point to each cluster center
centroids = []

for cluster_id in range(3):
    cluster_points = bert_embeddings_scaled[agg_clust.labels_ == cluster_id]
    centroid = np.mean(cluster_points, axis=0)
    centroids.append(centroid)
    
centroids = np.array(centroids)

# Computing distances to cluster centers
X_hierarchical_features = euclidean_distances(bert_embeddings_scaled, centroids)
```

### Exercise 1(j) (10 points)

Train and evaluate three `LogisticRegression` models as follows:

- Using `kmeans` features as input.
- Using `hierarchical` features as input.
- Using bert embedding features as input.
- Combining `kmeans` + bert embeddings features as input.
- Combining `hierarchical` + bert embeddings features as input.
- Combining `kmeans` + `hierarchical` + bert embeddings features as input.

Evaluate the three models over a five-fold cross validation strategy.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# Define the target variable
y = spam_df['label'].map({'ham': 0, 'spam': 1})

feature_sets = {
    'kmeans': X_kmeans_features,
    'hierarchical': X_hierarchical_features,
    'bert': bert_embeddings,
    'kmeans_bert': np.hstack((X_kmeans_features, bert_embeddings)),
    'hierarchical_bert': np.hstack((X_hierarchical_features, bert_embeddings)),
    'kmeans_hierarchical_bert': np.hstack((X_kmeans_features, X_hierarchical_features, bert_embeddings)),  
}

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Train and evaluate Logistic Regression models
model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000, random_state=42))

results = {}
for name, x in feature_sets.items():
    scores = cross_val_score(model, x, y, cv=skf, scoring='accuracy')
    results[name] = scores
    print(f"Model: {name} - Accuracy: {np.mean(scores):.4f} ± {np.std(scores):.4f}")
    print(f"Individual fold scores: {scores}")
    
# Finding the best performing model
best_model = max(results, key=lambda k: np.mean(results[k]))
print(f"\n Best performing model: {best_model} with accuracy {np.mean(results[best_model]):.4f}")
```
