### Exercise 2(c) (3 points)

Remove the number from `sample_text`. Consider the `remove_numbers` function to remove the numbers.

```
def remove_numbers(text):
    return ''.join([char for char in text if not char.isdigit()])

# Test the function
sample_text = "My phone number is 12345."
print(remove_numbers(sample_text))

```

````python
def remove_numbers(text):
    return ''.join([char for char in text if not char.isdigit()])

sample_text = remove_numbers(sample_text)
sample_text
````
